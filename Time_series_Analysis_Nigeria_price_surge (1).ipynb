```python
# importing the dataset
import pandas as pd
import numpy as np

data=pd.read_csv("/content/1960_onwards.csv")
data
```

```python
# converting the year column
data['Year'] = pd.to_datetime(data['Year'], format='%Y')
```

```python
# Time series for GDP growth (annual %)
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# Set 'Year' column as the index
data.set_index('Year', inplace=True)

# Select the GDP growth (annual %) column
gdp_growth = data['GDP growth (annual %)']

## Exploratory Data Analysis (EDA)

# Summary statistics
print("Summary Statistics:")
print(gdp_growth.describe())

# Plot the time series
plt.figure(figsize=(14, 7))
plt.plot(gdp_growth.index, gdp_growth, marker='o', linestyle='-')
plt.title('Time Series Analysis of GDP Growth (annual %)')
plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.grid(True)
plt.show()

# Plotting histogram
plt.figure(figsize=(10, 6))
sns.histplot(gdp_growth, bins=20, kde=True)
plt.title('Distribution of GDP Growth (annual %)')
plt.xlabel('GDP Growth (annual %)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plotting boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(y=gdp_growth)
plt.title('Boxplot of GDP Growth (annual %)')
plt.ylabel('GDP Growth (annual %)')
plt.grid(True)
plt.show()

# Plotting rolling mean and standard deviation
rolling_mean = gdp_growth.rolling(window=12).mean()
rolling_std = gdp_growth.rolling(window=12).std()

plt.figure(figsize=(14, 7))
plt.plot(gdp_growth.index, gdp_growth, label='Original')
plt.plot(rolling_mean, label='Rolling Mean')
plt.plot(rolling_std, label='Rolling Std Dev')
plt.title('Rolling Mean & Standard Deviation of GDP Growth (annual %)')
plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.legend()
plt.grid(True)
plt.show()

# Checking for stationarity with Augmented Dickey-Fuller test
adf_test = adfuller(gdp_growth)
print("\nResults of Augmented Dickey-Fuller Test:")
adf_results = pd.Series(adf_test[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
for key, value in adf_test[4].items():
    adf_results[f'Critical Value ({key})'] = value
print(adf_results)

## Time Series Analysis

# Perform seasonal decomposition to understand trend, seasonal, and residual components
decomposition = seasonal_decompose(gdp_growth, model='additive', period=1)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the components
plt.figure(figsize=(14, 10))
plt.subplot(411)
plt.plot(gdp_growth, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

# SARIMA model fitting and forecasting

model = SARIMAX(gdp_growth, order=(1, 1, 1), seasonal_order=(0, 1, 1, 12))
results = model.fit()

# Forecasting
forecast_periods = 12  # Example: forecasting for 12 periods (months/years)
forecast = results.get_forecast(steps=forecast_periods)

# Confidence intervals for the forecast
forecast_ci = forecast.conf_int()

# Plotting the forecasted values
plt.figure(figsize=(14, 7))
plt.plot(gdp_growth.index, gdp_growth, label='Actual')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, color='red', label='Forecast')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.2)
plt.title('GDP Growth (annual %) Forecast using SARIMA')
plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
